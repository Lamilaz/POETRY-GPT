```python
import pandas as pd
import kagglehub
path = kagglehub.dataset_download("carlosgdcj/genius-song-lyrics-with-language-information")
df = pd.read_csv(fr"{path}\song_lyrics.csv")
df_fr_precise = df[df["language_ft"].str.contains(r"\bfr\b", regex=True, na=False)]
df_fr_precise.to_csv("song_lyrics_fr.csv", index=False)
```
